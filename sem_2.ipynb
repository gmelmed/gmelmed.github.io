```python
import pandas as pd
import numpy as np
```

```python
# read in the consumer trends data
# the sheet called 02CN
# lines 8-35

ct = pd.read_excel('consumertrends.xlsx', sheet_name='02CN', skiprows=7, nrows=27)
ct
```

```python
# rename cols
ct.columns
```

```python
ct.rename(columns={
    "CDID identifier code": "id",
    "UUHX": "spirits",
    "ADFN": "tobacco"
}, inplace=True)
```

```python
ct = ct[['id', 'spirits', 'tobacco']]
ct
```

```python
# melt
ct = ct.melt(id_vars=['id'], var_name='category', value_name='value')
ct
```

```python
ct.columns = ['year', 'category', 'value']
```

```python
ct.to_csv('consumertrends.csv', index=False)
```
