```python
import pandas as pd
import numpy as np
```

```python
# read in the consumer trends data
# the sheet called 02CN
# lines 8-35

ct = pd.read_excel('consumertrends.xlsx', sheet_name='02CN', skiprows=7, nrows=27)
ct
```

```python
# rename cols
ct.columns
```

```python
ct.columns = [
    'id',
    'Alcohol & Tobacco',
    'Alcoholic Beverages',
    'Spirits',
    'Wines, cider and perry',
    'Beer',
    'Tobacco',
    'Narcotics']
```

```python
# melt
ct = ct.melt(id_vars=['id'], var_name='category', value_name='value')
ct
```

```python
ct.rename(columns={'id': 'year'}, inplace=True)
```

```python
ct.to_csv('consumertrends.csv', index=False)
```
