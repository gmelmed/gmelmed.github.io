```python
import pandas as pd
```

```python
df = pd.read_csv('data/lifeex.csv')

df
```

```python
# drop rows with fips null
df = df.dropna(subset=['fips'])

#  convert fips from float to int
df['fips'] = df['fips'].astype(int)
```

```python
df
```

```python
# only county
df = df[df['fips'] <= 1000]
df
```

```python
df[df['location_name'].str.contains('Louisiana')]
```

```python
# rename fips to id
df.rename(columns={'fips': 'id'}, inplace=True)
df
```

```python
df.to_csv('data/lifeex_counties.csv', index=False)
```
