```python
import pandas as pd
```

```python
df = pd.read_csv('data/lifeex.csv')

df
```

```python
# drop rows with fips null
df = df.dropna(subset=['fips'])

#  convert fips from float to int
df['fips'] = df['fips'].astype(int)
```

```python
df
```

```python
# only county
df = df[df['location_name'].str.contains('County')]
df
```

```python
df.to_csv('data/lifeex.csv', index=False)
```
