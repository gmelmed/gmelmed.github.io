I. Loading in prices data

```python
import pandas as pd
```

```python
# Add a progress bar to the reading process
prices_df = pd.read_csv('https://eco-prices-scrapes.s3.eu-west-2.amazonaws.com/teaching/davies_price_data/db_prices.csv')

prices_df
```

```python
prices_df
```

```python
items_df = pd.read_csv('https://eco-prices-scrapes.s3.eu-west-2.amazonaws.com/teaching/davies_price_data/db_item_clean.csv')

items_df
```

```python
# load in regions data
region_df = pd.read_csv("https://eco-prices-scrapes.s3.eu-west-2.amazonaws.com/teaching/davies_price_data/db_region.csv")

region_df
```

```python
# merge items and prices

full_df = pd.merge(prices_df, items_df[['item_id', 'description']], on='item_id') # merge prices and items

full_df

```

```python
region_codes = {
    2.0: "London",
    3.0: "South East",
    4.0: "South West",
    5.0: "East Anglia",
    6.0: "East Midlands",
    7.0: "West Midlands",
    8.0: "Yorkshire & Humber",
    9.0: "North West",
    10.0: "North",
    11.0: "Wales",
    12.0: "Scotland",
    13.0: "Northern Ireland"
}
```

```python
# add the region and country cols from the regions_df

full_df['region_name'] = full_df['region'].map(region_codes)

full_df

```

```python
items_df[items_df['description'].str.contains('hotel', case=False)]
```

```python
# restrict to only hotels

hotels_df = full_df[full_df['item_id'] == 640406]
```

```python
hotels_df
```

```python
# change quote_date from float yyyymm to datetime

hotels_df['quote_date'] = pd.to_datetime(hotels_df['quote_date'], format='%Y%m')

hotels_df
```
