```python
import pandas as pd
```

```python
# read in wide dataset
rent = pd.read_csv('us_rent_data_wide.csv')
rent
```

```python
rent = rent[rent['bed_size'] == '2br']
rent
```

```python
# Identify columns starting with '20'
date_cols = [col for col in rent.columns if str(col).startswith('20')]

# Convert those columns to datetime for easier manipulation
date_col_mapping = {col: pd.to_datetime(col, format='%Y_%m') for col in date_cols}
datetime_cols = pd.Series(date_col_mapping)

# Update the DataFrame to use datetime columns as column names temporarily
temp_rent = rent[date_cols].copy()
temp_rent.columns = datetime_cols.values

# Identify the most recent, one year before, and five years before columns
most_recent = temp_rent.columns.max()
one_year_prior = most_recent - pd.DateOffset(years=1)
five_years_prior = most_recent - pd.DateOffset(years=5)

# Ensure the columns exist in the dataset
valid_columns = temp_rent.columns
one_year_prior = one_year_prior if one_year_prior in valid_columns else None
five_years_prior = five_years_prior if five_years_prior in valid_columns else None

# Extract the values
rent['most_recent_value'] = temp_rent[most_recent]
rent['one_year_prior_value'] = temp_rent[one_year_prior] if one_year_prior else None
rent['five_years_prior_value'] = temp_rent[five_years_prior] if five_years_prior else None

# Calculate the percentage changes
rent['1_year_pct_change'] = (
    (rent['most_recent_value'] - rent['one_year_prior_value']) / rent['one_year_prior_value'] * 100
    if one_year_prior else None
)
rent['5_year_pct_change'] = (
    (rent['most_recent_value'] - rent['five_years_prior_value']) / rent['five_years_prior_value'] * 100
    if five_years_prior else None
)

rent
```

```python
import json

# Filter rows where 'location_type' is 'City'
city_data = rent[rent['location_type'] == 'City']

# Sort by 'population' in descending order and select the top 20
top_cities = city_data.nlargest(20, 'population')

# Extract 'location_name' values and convert to a JSON format
top_cities_json = json.dumps(top_cities['location_name'].tolist())

# Save the JSON to a file
with open('top_cities.json', 'w') as json_file:
    json_file.write(top_cities_json)
```
