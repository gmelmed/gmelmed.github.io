```python
import pandas as pd
```

```python
# Read in ZORI metro data
zori_metro_wide = pd.read_csv('zori_metro_wide.csv')
zori_metro_wide
```

```python
zori_metro_wide.columns
```

```python
index_cols = ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName']
```

```python
# melt df
zori_metro_long = pd.melt(
    zori_metro_wide,
    id_vars = index_cols,
    value_vars = [col for col in zori_metro_wide.columns if col not in index_cols],
    var_name = "date",
    value_name = "zori"
)

zori_metro_long
```

```python
zori_metro_long['date'] = pd.to_datetime(zori_metro_long['date'])

zori_metro_long
```

```python
cities = pd.read_csv('uscities.csv')
cities
```

```python
# create an aligned name column

# create a name column with the City, StateID format
cities["name"] = cities["city"] + ", " + cities["state_id"]

# rename name column in zori df
zori_metro_long.rename(
    columns={
        "RegionName":"name"
    },
    inplace=True
)
```

```python
# checking if the values are the same

# Step 1: Find common values
common_values = set(zori_metro_long["name"]).intersection(set(cities["name"]))

# Step 2: Length of common values
print(f"Number of common values: {len(common_values)}")

# Step 3: List of 20 largest cities (example list, replace with your own as needed)
largest_cities = [
    "New York, NY", "Los Angeles, CA", "Chicago, IL", "Houston, TX",
    "Phoenix, AZ", "Philadelphia, PA", "San Antonio, TX", "San Diego, CA",
    "Dallas, TX", "San Jose, CA", "Austin, TX", "Jacksonville, FL",
    "Fort Worth, TX", "Columbus, OH", "Charlotte, NC", "San Francisco, CA",
    "Indianapolis, IN", "Seattle, WA", "Denver, CO", "Washington, DC"
]

# Step 4: Check which of the 20 largest cities are in the common values
largest_cities_in_common = set(largest_cities).intersection(common_values)
print(f"Largest cities in common values: {largest_cities_in_common}")
print(f"Number of largest cities in common: {len(largest_cities_in_common)}")

# Step 5: Check if all 20 largest cities are present
if len(largest_cities_in_common) == len(largest_cities):
    print("All 20 largest cities are present in the common values.")
else:
    print(f"Some largest cities are missing. Missing cities: {set(largest_cities) - largest_cities_in_common}")
```

```python
# Merge the two DataFrames
zori_metro_long = zori_metro_long.merge(
    cities[['name', 'population', 'lat', 'lng']],  # Select only relevant columns from `cities`
    # may have to deal with population another way later
    # it's city population, not msa population.
    on='name',
    how='left'
)

zori_metro_long
```
