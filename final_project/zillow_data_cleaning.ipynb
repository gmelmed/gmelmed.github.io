```python
import pandas as pd
```

```python
# Read in ZORI metro data
zori_metro_wide = pd.read_csv('zori_metro_wide.csv')
zori_metro_wide
```

```python
# drop rows with no population value
zori_metro_wide = zori_metro_wide.dropna(subset=['pop_2023'])
```

```python
# convert pop_2023 to numeric
zori_metro_wide['pop_2023'] = pd.to_numeric(zori_metro_wide['pop_2023'].str.replace(',', ''), errors='coerce')

# sort by population
zori_metro_wide.sort_values(
    by = 'pop_2023',
    ascending = False,
    inplace = True
)

zori_metro_wide
```

```python
import json

# Filter rows where 'location_type' is 'City'
metro_data = zori_metro_wide[zori_metro_wide['RegionType'] == 'msa']

# Sort by 'population' in descending order and select the top 20
top_metros = metro_data.nlargest(50, 'pop_2023')

# Extract 'location_name' values and convert to a JSON format
top_metros_json = json.dumps(top_metros['RegionName'].tolist())

# Save the JSON to a file
with open('top_metros.json', 'w') as json_file:
    json_file.write(top_metros_json)
```

## Melting df

```python
zori_metro_wide.columns
```

```python
index_cols = ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'pop_2023']
```

```python
# melt df
zori_metro_long = pd.melt(
    zori_metro_wide,
    id_vars = index_cols,
    value_vars = [col for col in zori_metro_wide.columns if col not in index_cols],
    var_name = "date",
    value_name = "zori"
)

zori_metro_long
```

```python
zori_metro_long['date'] = pd.to_datetime(zori_metro_long['date'])

zori_metro_long
```

```python
cities = pd.read_csv('uscities.csv')
cities
```

```python
# create an aligned name column

# create a name column with the City, StateID format
cities["name"] = cities["city"] + ", " + cities["state_id"]

# rename name column in zori df
zori_metro_long.rename(
    columns={
        "RegionName":"name"
    },
    inplace=True
)
```

```python
cities['name']
```

```python
# Merge the two DataFrames
zori_metro_long = zori_metro_long.merge(
    cities[['name', 'lat', 'lng']],  # Select only relevant columns from `cities`
    on='name',
    how='left'
)

zori_metro_long
```

```python
zori_metro_long
```

## Market heat index

```python
# Read in market temp
heat_metro_wide = pd.read_csv('zillow_temp_wide.csv')
heat_metro_wide

# drop rows with no population value
heat_metro_wide = heat_metro_wide.dropna(subset=['pop_2023'])

# convert pop_2023 to numeric
heat_metro_wide['pop_2023'] = pd.to_numeric(heat_metro_wide['pop_2023'].str.replace(',', ''), errors='coerce')

# sort by population
heat_metro_wide.sort_values(
    by = 'pop_2023',
    ascending = False,
    inplace = True
)

# melt df
heat_metro_long = pd.melt(
    heat_metro_wide,
    id_vars = index_cols,
    value_vars = [col for col in heat_metro_wide.columns if col not in index_cols],
    var_name = "date",
    value_name = "heat"
)

# convert date to datetime 
heat_metro_long['date'] = pd.to_datetime(heat_metro_long['date'])

# rename name column in heat df
heat_metro_long.rename(
    columns={
        "RegionName":"name"
    },
    inplace=True
)

# Merge the two DataFrames
heat_metro_long = heat_metro_long.merge(
    cities[['name', 'lat', 'lng']],  # Select only relevant columns from `cities`
    on='name',
    how='left'
)

heat_metro_long
```

```python
heat_zori_metro = heat_metro_long.merge(
    zori_metro_long,
    on=[col for col in heat_metro_long.columns if col not in ['heat']],
    suffixes=('_heat', '_zori')
)

heat_zori_metro
```

```python
heat_zori_metro.to_csv('heat_zori_metro_clean.csv')
```
