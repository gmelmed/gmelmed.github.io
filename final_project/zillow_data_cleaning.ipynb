## Market heat index

```python
# Read in market temp
heat_metro_wide = pd.read_csv('data/zillow_temp_wide.csv')
heat_metro_wide

# drop rows with no population value
heat_metro_wide = heat_metro_wide.dropna(subset=['pop_2023'])

# convert pop_2023 to numeric
heat_metro_wide['pop_2023'] = pd.to_numeric(heat_metro_wide['pop_2023'].str.replace(',', ''), errors='coerce')

# sort by population
heat_metro_wide.sort_values(
    by = 'pop_2023',
    ascending = False,
    inplace = True
)

# melt df
heat_metro_long = pd.melt(
    heat_metro_wide,
    id_vars = index_cols,
    value_vars = [col for col in heat_metro_wide.columns if col not in index_cols],
    var_name = "date",
    value_name = "heat"
)

# convert date to datetime 
heat_metro_long['date'] = pd.to_datetime(heat_metro_long['date'])

# rename name column in heat df
heat_metro_long.rename(
    columns={
        "RegionName":"name"
    },
    inplace=True
)

# Merge the two DataFrames
heat_metro_long = heat_metro_long.merge(
    cities[['name', 'lat', 'lng']],  # Select only relevant columns from `cities`
    on='name',
    how='left'
)

heat_metro_long
```

```python
heat_zori_metro = heat_metro_long.merge(
    zori_metro_long,
    on=[col for col in heat_metro_long.columns if col not in ['heat']],
    suffixes=('_heat', '_zori')
)

heat_zori_metro
```

```python
heat_zori_metro.to_csv('data/heat_zori_metro_clean.csv')
```

```python
# Step 1: Find the most recent date in the dataset
most_recent_date = heat_zori_metro['date'].max()

# Step 2: Calculate the 'most_recent_heat' and 'most_recent_zori' for each metro area
most_recent_data = heat_zori_metro[heat_zori_metro['date'] == most_recent_date]

# Step 3: Initialize the result DataFrame with 'name' and 'pop_2023' columns
result_df = most_recent_data[['name','pop_2023','lat','lng']].copy()
result_df['most_recent_heat'] = most_recent_data['heat']
result_df['most_recent_zori'] = most_recent_data['zori']
result_df['most_recent_date'] = most_recent_date

# Step 4: Calculate year-over-year percent changes for 'heat' and 'zori'
for years in range(1, 6):  # Loop for 1-year to 5-year changes
    # Calculate the date N years before the most recent date
    previous_date = most_recent_date - pd.DateOffset(years=years)

    # Filter data for the previous date for each metro area
    previous_data = heat_zori_metro[heat_zori_metro['date'] == previous_date]
    previous_data = previous_data.set_index('name')  # Set 'name' as index for easy lookup

    # Calculate percent change for 'heat' and 'zori'
    result_df[f'heat_{years}yc'] = (
        (result_df['most_recent_heat'] - result_df['name'].map(previous_data['heat'])) /
        result_df['name'].map(previous_data['heat'])
    ) * 100

    result_df[f'zori_{years}yc'] = (
        (result_df['most_recent_zori'] - result_df['name'].map(previous_data['zori'])) /
        result_df['name'].map(previous_data['zori'])
    ) * 100
```

```python
result_df
```

```python
# get rid of the row where name == 'United States'
result_df = result_df[result_df['name'] != 'United States']
result_df
```

```python
result_df.to_csv('data/heat_zori_change.csv', index=False)
```

## population centroids

```python
import pandas as pd
```

```python
county_centroids_url = "https://www2.census.gov/geo/docs/reference/cenpop2020/county/CenPop2020_Mean_CO.txt"

county_centroids = pd.read_csv(county_centroids_url)

county_centroids.to_csv('data/county_centroids.csv', index=False)
```

```python
county_centroids.dtypes
```

```python
# create fips column
county_centroids['fips'] = county_centroids['STATEFP'] * 1000 + county_centroids['COUNTYFP']

county_centroids
```

```python
county_centroids[county_centroids['STNAME'] == 'Connecticut']
```

```python
# read in population data
pop_est = pd.read_excel('data/PopulationEstimates.xlsx', skiprows=4,)

pop_est
```

```python
pop_est[pop_est['State'] == 'CT']
```

```python
pop_2023 = pop_est[['FIPStxt', 'POP_ESTIMATE_2023']]
pop_2023
```

```python
# unified col names
pop_2023.columns = ['fips', 'pop_2023']
```

```python
# merge county centroids with population data
county_centroids = county_centroids.merge(
    pop_2023,
    on='fips',
    how='left'
)

# drop unnecessary columns
county_centroids.drop(
    columns=['STATEFP', 'COUNTYFP', 'POPULATION', 'COUNAME', 'STNAME'],
    inplace=True
)

county_centroids
```

```python
# rename columns
county_centroids.columns = ['lat', 'lng', 'fips', 'pop_2023']

county_centroids
```

```python
# form a fips col for zori_county_wide
zori_county_wide['fips'] = zori_county_wide['StateCodeFIPS'] * 1000 + zori_county_wide['MunicipalCodeFIPS']

# form a name col for zori_county_wide
zori_county_wide['name'] = zori_county_wide['RegionName'] + ', ' + zori_county_wide['State']

zori_county_wide
```

```python
# merge with population centroids
zori_county_wide = zori_county_wide.merge(
    county_centroids,
    on = 'fips',
    how='left'
)

zori_county_wide
```

```python
# create list of index cols

index_cols = ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName',
       'State', 'Metro', 'StateCodeFIPS', 'MunicipalCodeFIPS', 'fips', 'name', 'lat', 'lng', 'pop_2023']
```

```python
# melt
zori_county_long = pd.melt(
    zori_county_wide,
    id_vars = index_cols,
    value_vars = [col for col in zori_county_wide.columns if col not in index_cols],
    var_name = "date",
    value_name = "zori"
)

zori_county_long
```

```python
# keep only necessary cols

zori_county_long_clean = zori_county_long[['name','RegionType', 'fips', 'lat', 'lng', 'pop_2023', 'date', 'zori']]

# rename RegionType to type  
zori_county_long_clean.rename(
    columns={'RegionType':'type'},
    inplace=True
)

zori_county_long_clean
```

```python
# convert date to datetime
zori_county_long_clean['date'] = pd.to_datetime(zori_county_long_clean['date'])

zori_county_long_clean.dtypes
```

```python
zori_county_long_clean
```

```python
# create 1-5 year percent change columns
for years in range(1, 6):
    zori_county_long_clean[f'zori_{years}yc'] = zori_county_long_clean.groupby('name')['zori'].pct_change(12 * years) * 100

zori_county_long_clean
```

```python
# save to csv
zori_county_long_clean.to_csv('data/zori_county_long_clean.csv', index=False)
```

```python
# create a version of the df with only the most recent data

# find the most recent date
most_recent_date = zori_county_long_clean['date'].max()
# filter for most recent date
zori_county_most_recent_clean = zori_county_long_clean[zori_county_long_clean['date'] == most_recent_date]

# drop cols
zori_county_most_recent_clean.drop(columns=[
    'type',
    'zori_2yc',
    'zori_3yc',
    'zori_4yc'
], 
inplace=True
)

zori_county_most_recent_clean
```

```python
# save to csv
zori_county_most_recent_clean.to_csv('data/zori_county_most_recent_clean.csv', index=False)
```
