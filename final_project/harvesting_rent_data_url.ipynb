```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
```

```python

cookies = {
    'geofence_bypass': 'true'
}

response = requests.get('https://www.apartmentlist.com/research/category/data-rent-estimates', cookies=cookies)


with open("temp.html", "w") as f:
    f.write(response.text)
```

```python
with open("temp.html", "r") as f:
    html = f.read()

soup = BeautifulSoup(html, 'html.parser')

# Find the <script> tag with id = __NEXT_DATA__
script_tag = soup.find("script", {"id": "__NEXT_DATA__"})

page_data = script_tag.text
page_data = json.loads(page_data)

with open("temp.json", "w") as f:
    f.write(json.dumps(page_data, indent=4))

file_url = page_data["props"]["pageProps"]["component"]["searchResults"][0]["fields"]["downloadableAssets"][1]["fields"]["attachment"]["fields"]["file"]["url"]

file_url
```

```python
url = "https:" + file_url
```

```python
url
```

```python
df = pd.read_csv(url)
df
```

```python
us_rent = df[df["location_type"] == "National"]
us_rent
```

```python
# Filter the columns: keep "bed_size" and date cols
columns_to_keep = ['bed_size'] + [col for col in us_rent.columns if '20' in col]

# Create the reduced DataFrame
us_rent = us_rent[columns_to_keep]
us_rent
```

```python
us_rent_long = pd.melt(
    us_rent,
    id_vars = ['bed_size'],
    value_vars = [col for col in us_rent.columns if '20' in col],
    var_name = "date",
    value_name = "rent"
)

us_rent_long
```

```python
us_rent_long.to_csv('us_rent_clean.csv', index=False)
```

```python
# from the census bureau

# counties
counties_popcenter_url = 'https://www2.census.gov/geo/docs/reference/cenpop2020/county/CenPop2020_Mean_CO.txt'

# states
states_popcenter_url = 'https://www2.census.gov/geo/docs/reference/cenpop2020/CenPop2020_Mean_ST.txt'
```

```python
fips_points = pd.read_csv('https://www2.census.gov/geo/docs/reference/cenpop2020/county/CenPop2020_Mean_CO.txt')
fips_points
```

```python
# dictionary to map state names to their ID
state_to_id = {
    "Alabama": "AL", "Alaska": "AK", "Arizona": "AZ", "Arkansas": "AR", "California": "CA",
    "Colorado": "CO", "Connecticut": "CT", "Delaware": "DE", "Florida": "FL", "Georgia": "GA",
    "Hawaii": "HI", "Idaho": "ID", "Illinois": "IL", "Indiana": "IN", "Iowa": "IA",
    "Kansas": "KS", "Kentucky": "KY", "Louisiana": "LA", "Maine": "ME", "Maryland": "MD",
    "Massachusetts": "MA", "Michigan": "MI", "Minnesota": "MN", "Mississippi": "MS", "Missouri": "MO",
    "Montana": "MT", "Nebraska": "NE", "Nevada": "NV", "New Hampshire": "NH", "New Jersey": "NJ",
    "New Mexico": "NM", "New York": "NY", "North Carolina": "NC", "North Dakota": "ND", "Ohio": "OH",
    "Oklahoma": "OK", "Oregon": "OR", "Pennsylvania": "PA", "Rhode Island": "RI", "South Carolina": "SC",
    "South Dakota": "SD", "Tennessee": "TN", "Texas": "TX", "Utah": "UT", "Vermont": "VT",
    "Virginia": "VA", "Washington": "WA", "West Virginia": "WV", "Wisconsin": "WI", "Wyoming": "WY",
    "District of Columbia": "DC", "Puerto Rico" : "PR"
}
```

```python
# map states to id
fips_points["state_code"] = fips_points["STNAME"].map(state_to_id)
fips_points
```

```python
# create harmonized "name" col
fips_points["name"] = fips_points['COUNAME'] + " County, " + fips_points["state_code"]
fips_points
```

```python
# harmonize col names 
fips_points = fips_points[["name", "LATITUDE", "LONGITUDE"]]
fips_points.columns = ["name", "lat", "lng"]
fips_points
```

```python
# merge
counties_rent_points = counties_rent.merge(fips_points, left_on="location_name", right_on="name", how="left")
counties_rent_points
```

```python
# read in states population centroids:
states_points = pd.read_csv(states_popcenter_url)
states_points
```

```python
states_points.columns
```

```python
# harmonize column names
states_points = states_points[['STNAME', 'LATITUDE', 'LONGITUDE']]
states_points.columns = ['name', 'lat', 'lng']
```

```python
# merge
states_rent_points = states_rent.merge(states_points, left_on="location_name", right_on="name", how="left")
states_rent_points
```

```python
# concatonate

# check that they all have the same cols
# Extract the columns as sets
states_columns = set(states_rent_points.columns)
cities_columns = set(cities_rent_points.columns)
counties_columns = set(counties_rent_points.columns)

# Compare the sets
if states_columns == cities_columns == counties_columns:
    print("All three DataFrames have the same columns.")
else:
    print("The columns differ between the DataFrames.")
    # Show differences
    print("Columns in states_rent_points but not in the others:", states_columns - cities_columns - counties_columns)
    print("Columns in cities_rent_points but not in the others:", cities_columns - states_columns - counties_columns)
    print("Columns in counties_rent_points but not in the others:", counties_columns - states_columns - cities_columns)
```

```python
# Concatenate the DataFrames
rent_wide = pd.concat([states_rent_points, cities_rent_points, counties_rent_points], ignore_index=True)

rent_wide
```

```python
rent_wide.drop(
    columns = "name",
    inplace = True
)
rent_wide
```

```python
rent_wide
```

```python
# set id vars
id_vars = ['location_name', 'location_type', 'location_fips_code', 'population',
           'state', 'county', 'metro', 'lat', 'lng', 'bed_size']

value_vars = [col for col in rent_wide.columns if col not in id_vars]
```

```python
# melt the df
rent_long = pd.melt(rent_wide, id_vars=id_vars, value_vars=value_vars,
                    var_name='date', value_name='rent')

rent_long
```

```python
# convert 'date' column to datetime format
rent_long['date'] = pd.to_datetime(rent_long['date'], format='%Y_%m')
rent_long
```

```python
rent_long.to_csv('us_rent_points_clean.csv')
```
