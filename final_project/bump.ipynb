```python
import pandas as pd
```

```python
# read in rent date
rent = pd.read_csv('zori_metro_long_clean.csv')
```

```python
# ger rid of "United States" 
rent = rent[rent['name'] != 'United States']
```

```python
rent.dtypes
```

```python
# convert SizeRank to string
rent['SizeRank'] = rent['SizeRank'].astype(str)
# strip the whitespace from the integer column SizeRank
rent['SizeRank'] = rent['SizeRank'].str.strip()
#convert the column to integer
rent['SizeRank'] = rent['SizeRank'].astype(int)
```

```python
# show the unique values in the name SizeRank
rent['SizeRank'].unique()
```

```python

rent = rent[rent['SizeRank'] <= 50]
rent
```

```python
rent['rank'] = rent.groupby('date')['zori'].rank(ascending=False, method='first')

rent
```
