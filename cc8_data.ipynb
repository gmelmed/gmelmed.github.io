```python
items_df = pd.read_csv('https://eco-prices-scrapes.s3.eu-west-2.amazonaws.com/teaching/davies_price_data/db_item_clean.csv')
```

```python
from tqdm import tqdm

# Add a progress bar to the reading process
prices_df = pd.read_csv('https://eco-prices-scrapes.s3.eu-west-2.amazonaws.com/teaching/davies_price_data/db_prices.csv', iterator=True, chunksize=1000)

prices_df = pd.concat([chunk for chunk in tqdm(prices_df, desc="Loading prices data")])
print(f"Loaded {len(prices_df)} rows of prices data")
```

```python
items_df.columns
```

```python
full_df = pd.merge(prices_df, items_df[['item_id', 'description']], on='item_id') # merge prices and items
full_df
```

```python
full_df['quote_date'] = pd.to_datetime(full_df['quote_date'], format='%Y%m') # convert quote_date to datetime, format is yearmonth
full_df
```

```python
# filter for items with keyword in the description
items_df[items_df['description'].str.contains('jacket', case=False)]
```

```python
suit_items = {
    510104: "Jacket",
    510117: "Formal Trousers",
    510413: "Socks",
    510109: "Formal Shirt",
    510501: "Formal Shoes",
    510419: "Tie"
}
```

```python
items_df[items_df['item_id'].isin(suit_items.keys())]
```

```python
# filter full_df to only include suit items
suit_df = full_df[full_df['item_id'].isin(suit_items.keys())] 

suit_df
```

```python
suit_avg = suit_df.groupby(['quote_date', 'description', 'item_id']).agg({'price': 'mean'}).reset_index()

suit_avg
```

```python
# Group by quote_date and filter dates where all item_ids are present
complete_dates = suit_avg.groupby('quote_date').filter(lambda x: set(suit_items.keys()).issubset(set(x['item_id'])))

# Find the earliest date
earliest_complete_date = complete_dates['quote_date'].min()

# Show the earliest date
earliest_complete_date
```

```python
# filter for data after the earliest complete date
suit_avg = suit_avg[suit_avg['quote_date'] >= earliest_complete_date]
```

```python
# add a item_name col with the values from suit_items to the description column
suit_avg['item_name'] = suit_avg['item_id'].map(suit_items)

suit_avg
```
