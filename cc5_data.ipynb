```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
advanced_stats_url = "https://www.basketball-reference.com/leagues/NBA_2025_advanced.html"

data = requests.get(advanced_stats_url)

3 data.text
```

```python
# Parse the HTML as a string
soup = BeautifulSoup(data.text, 'html.parser') 
```

```python
advanced_stats_table = soup.find(id="advanced")
```

```python
nba = pd.read_html(str(advanced_stats_table))[0]
nba
```

```python
nba = pd.read_csv("data/nba.csv")

nba
```

```python
average_vorp_by_team = nba.groupby('Team')['VORP'].mean()

average_vorp_by_team.columns = ['Team', 'Average VORP']

# sort the teams by average VORP
average_vorp_by_team = average_vorp_by_team.sort_values(ascending=False)

average_vorp_by_team
```

```python
# Create a new DataFrame to store the results
team_stats = pd.DataFrame(columns=['id', 'avg_VORP', 'best_player', 'best_VORP', 'avg_VORP_without_best'])

# Iterate through each team
for team in nba['Team'].unique():
    team_data = nba[nba['Team'] == team]
    average_vorp = team_data['VORP'].mean()
    best_player = team_data.loc[team_data['VORP'].idxmax()]
    best_player_vorp = best_player['VORP']
    average_vorp_without_best = team_data[team_data['Player'] != best_player['Player']]['VORP'].mean()
    
    team_stats = team_stats.append({
        'id': team,
        'avg_VORP': average_vorp,
        'best_player': best_player['Player'],
        'best_VORP': best_player_vorp,
        'avg_VORP_without_best': average_vorp_without_best
    }, ignore_index=True)

team_stats = team_stats[team_stats['id'] != '2TM']

team_stats
```

```python
nba_teams = {
    "ATL": "Atlanta Hawks",
    "BOS": "Boston Celtics",
    "BRK": "Brooklyn Nets",
    "CHI": "Chicago Bulls",
    "CHO": "Charlotte Hornets",
    "CLE": "Cleveland Cavaliers",
    "DAL": "Dallas Mavericks",
    "DEN": "Denver Nuggets",
    "DET": "Detroit Pistons",
    "GSW": "Golden State Warriors",
    "HOU": "Houston Rockets",
    "IND": "Indiana Pacers",
    "LAC": "Los Angeles Clippers",
    "LAL": "Los Angeles Lakers",
    "MEM": "Memphis Grizzlies",
    "MIA": "Miami Heat",
    "MIL": "Milwaukee Bucks",
    "MIN": "Minnesota Timberwolves",
    "NOP": "New Orleans Pelicans",
    "NYK": "New York Knicks",
    "OKC": "Oklahoma City Thunder",
    "ORL": "Orlando Magic",
    "PHI": "Philadelphia 76ers",
    "PHO": "Phoenix Suns",
    "POR": "Portland Trail Blazers",
    "SAC": "Sacramento Kings",
    "SAS": "San Antonio Spurs",
    "TOR": "Toronto Raptors",
    "UTA": "Utah Jazz",
    "WAS": "Washington Wizards"
}

# create a new column with the full team name from the id col
team_stats['team'] = team_stats['id'].map(nba_teams)

team_stats
```

```python
nba_vorp = nba[['Player', 'Team', 'VORP']]

nba_vorp.columns = ['player', 'id', 'VORP']

# add a binary 'best' column that is 1 if the 'Player' value is in the 'best_player' column of team_stats
nba_vorp['best'] = nba_vorp['player'].isin(team_stats['best_player']).astype(int)

nba_vorp
```

```python
nba_vorp = nba_vorp[nba_vorp['id'] != '2TM']
```

```python
# Get team averages
team_avgs = nba_vorp.groupby('id')['VORP'].mean().reset_index()
team_avgs['player'] = team_avgs['id'] + ' average'
team_avgs['best'] = 0

# Get team averages excluding best players
no_best_avgs = nba_vorp[nba_vorp['best'] == 0].groupby('id')['VORP'].mean().reset_index()
no_best_avgs['player'] = no_best_avgs['id'] + ' average without best player'
no_best_avgs['best'] = 0

# Combine original df with new rows
nba_vorp = pd.concat([nba_vorp, team_avgs, no_best_avgs], ignore_index=True)

nba_vorp
```

```python
def categorize(row):
    if row['best'] == 1:
        return 'Best Player'
    elif 'average' in row['player'] and 'without' not in row['player']:
        return 'Average'
    elif 'average' in row['player'] and 'without' in row['player']:
        return 'Average without Best'
    else:
        return 'other'

nba_vorp['category'] = nba_vorp.apply(categorize, axis=1)

nba_vorp
```

```python
# add a surname column
nba_vorp['surname'] = nba_vorp['player'].str.split().str[1]

nba_vorp
```

```python
# add the team name to the DataFrame
nba_vorp['team'] = nba_vorp['id'].map(nba_teams)

nba_vorp
```

```python
nba_vorp.to_csv('data/vorp.csv', index=False)
```
