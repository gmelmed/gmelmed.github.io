```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
```

```python

cookies = {
    'geofence_bypass': 'true'
}

response = requests.get('https://www.apartmentlist.com/research/category/data-rent-estimates', cookies=cookies)


with open("temp.html", "w") as f:
    f.write(response.text)
```

```python
with open("temp.html", "r") as f:
    html = f.read()

soup = BeautifulSoup(html, 'html.parser')

# Find the <script> tag with id = __NEXT_DATA__
script_tag = soup.find("script", {"id": "__NEXT_DATA__"})

page_data = script_tag.text
page_data = json.loads(page_data)

with open("temp.json", "w") as f:
    f.write(json.dumps(page_data, indent=4))

file_url = page_data["props"]["pageProps"]["component"]["searchResults"][0]["fields"]["downloadableAssets"][1]["fields"]["attachment"]["fields"]["file"]["url"]

file_url
```

```python
url = "https:" + file_url
```

```python
df = pd.read_csv(url)
df
```
